```python
import pandas as pd

df = pd.read_csv("/content/used_cars_dataset.csv")
df
```

```python
df.fuel.unique()
```

```python
fuel_dic = {
    'Electric':0,
    'Petrol':1,
    'LPG':2,
    'Diesel':3,
    'CNG':4
}
df.fuel = df.fuel.map(fuel_dic)
```

```python
df.head(5)
```

```python
df.seller_type.unique()
```

```python
seller_dic = {
    'Individual': 0,
    'Dealer': 1,
    'Trustmark Dealer':2
}
df.seller_type = df.seller_type.map(seller_dic)
```

```python
df.head()
```

```python
df.transmission.unique()
```

```python
trans_dic = {
    'Automatic':0,
     'Manual':1
}
df.transmission = df.transmission.map(trans_dic)
```

```python
df.head()
```

```python
df.owner.unique()
```

```python
owner_dic = {
    'Third Owner':0,
    'First Owner':1,
    'Second Owner':2,
    'Fourth & Above Owner':3
}
df.owner = df.owner.map(owner_dic)
```

```python
df.head(3)
```

```python
df.is_roof.unique()
```

```python
roof_dic = {
    'No':0,
    'Yes':1
}
df.is_roof = df.is_roof.map(roof_dic)
```

```python
df.head()
```

```python
df.power_window.unique()
```

```python
window_dic = {
    'No':0,
    'Yes':1
}
df.power_window = df.power_window.map(window_dic)
```

```python
df.head()
```

```python
df.info()
```

```python
df.head(5)
```

```python
df.mileage = df.mileage.str.replace(" kmpl", "").astype(float)
```

```python
df.head()
```

```python
df.engine = df.engine.str.replace(" CC", "").astype(int)
```

```python
df.head()
```

```python
df.max_power = df.max_power.str.replace(" bhp", "").astype(float)
```

```python
df.head()
```

```python
df.drop(['name', 'torque'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop(['selling_price'], axis=1)
y = df.selling_price
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("MSE: ", mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred))
```

```python
from sklearn.svm import SVR
model2 = SVR()
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
print("MSE: ", mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor()
model3.fit(X_train, y_train)
```

```python
y_pred = model3.predict(X_test)
```

```python
print("MSE: ", mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
print("R2: ", r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
model4 = RandomForestRegressor()
model4.fit(X_train, y_train)
```
