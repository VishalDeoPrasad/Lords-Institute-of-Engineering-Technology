```python
import pandas as pd

df = pd.read_csv("/content/zomato_data.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.weather.unique()
```

```python
weather_dic = {
    'cloudy': 0,
    'rainy': 1,
    'sunny': 2
}
df.weather = df.weather.map(weather_dic)
```

```python
df.head()
```

```python
df.time_of_day.unique()
```

```python
time_dic = {
    'peak': 0,
    'off-peak': 1
}
df.time_of_day = df.time_of_day.map(time_dic)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.traffic_level.unique()
```

```python
traffic_dic = {
    'medium': 1,
    'high': 2,
    'low': 0
}
df.traffic_level = df.traffic_level.map(traffic_dic)
```

```python
df.head()
```

```python
df.area_type.unique()
```

```python
area_dic = {
    'urban':0,
    'suburban':1,
    'rural': 2
}
df.area_type = df.area_type.map(area_dic)
```

```python
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
x = df.drop('delivery_time_minutes', axis=1)
y = df.delivery_time_minutes
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_train, y_train)
```

```python
y_pred = model1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
lr_mse = mean_squared_error(y_test, y_pred)
lr_mse
```

```python
from sklearn.metrics import mean_absolute_error
lr_mae = mean_absolute_error(y_test, y_pred)
lr_mae
```

```python
from sklearn.metrics import r2_score
lr_r2 = r2_score(y_test, y_pred)
lr_r2
```

```python
from sklearn.svm import SVR
model2 = SVR()
model2.fit(x_train, y_train)
```

```python
y_pred = model2.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor

model3 = DecisionTreeRegressor()
model3.fit(x_train, y_train)
```

```python
y_pred = model3.predict(x_test)
y_pred
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

model4 = RandomForestRegressor()
model4.fit(x_train, y_train)
```

```python
y_pred = model4.predict(x_test)
y_pred
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

```python
import joblib
joblib.dump(model4, 'zomato_time_estimation_model.pkl')
```
