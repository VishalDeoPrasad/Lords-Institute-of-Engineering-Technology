# Matplotlib

```python
import matplotlib.pyplot as plt

year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
population = [12000, 13000, 14000, 16000, 16000, 17000, 18000, 19000, 20000]

plt.plot(year, population)
plt.xlabel('year')
plt.ylabel("Population")
```

```python
years = [2018, 2019, 2020, 2021, 2022, 2023]
sales = [25000, 30000, 28000, 35000, 40000, 45000]

plt.plot(years, sales)
plt.xlabel('Years')
plt.ylabel("Sales")
```

```python
size = ['S', 'M', 'L', 'XL']
qty = [23, 34, 12, 9]

plt.bar(size, qty)
```

```python
size = ['S', 'M', 'L', 'XL']
qty = [23, 34, 12, 9]

plt.barh(size, qty)
plt.title("Size Distribution")
plt.xlabel("Size")
plt.ylabel("Quantity")
```

```python
age = [30,40,60,50,30,40,5,6,54,45,7,43,34,6,16,18, 19]

plt.hist(age, bins=20);
```

```python
import numpy as np
data = np.random.randint(1, 100, 1000)

plt.hist(data);
```

```python
import numpy as np
import matplotlib.pyplot as plt

age  = np.random.randint(1, 100, 1000)
plt.hist(age);
```

```python
import numpy as np
import matplotlib.pyplot as plt

age = np.random.randint(1, 100, 1000)

plt.hist(age, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
X = [1,2,3,4,5]
Y = [10,8,4,2,1]

plt.scatter(X, Y)
```

```python
X = [1,2,3,4,5]
Y = [2,3,6,9,10]

plt.scatter(X, Y)
```

```python
size = ["S", "M", "L"]
qty = [20, 50, 10]

plt.pie(qty, labels=size, autopct="%1.2f%%")
plt.title("T-shirt Size Distribution");
```
